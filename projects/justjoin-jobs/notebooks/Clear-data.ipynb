# Dodajmy kilka nowych cech

Jak możemy zauważyć kolumna **employment_types** oraz **skills** zawierają dużo ciekawych danych, które warto wykorzystać.

```python
df.info()
```

```python
df['employment_types'] = df['employment_types'].apply(lambda x: eval(x.strip('[]')))
df['skills'] = df['skills'].apply(lambda x: eval(x.strip('[]')))
```

```python
df.head()
```

```python
def get_contract_type(data):
    if len(data) > 1:
        try:
            return data['type']
        except:
            return "More"
    return data['type']
```

```python
df['contract_type'] = df['employment_types'].apply(lambda x: get_contract_type(x))
```

```python
df['contract_type'].unique()
```

```python
df.head()
```

## Price

```python
def get_salary(data, label="from"):
    if len(data) > 1:
        try:
            return data['salary'][label]
        except:
            return "None"
    return data['salary'][label]
```

```python
df['salary_min'] = df['employment_types'].apply(lambda x: get_salary(x))
df['salary_max'] = df['employment_types'].apply(lambda x: get_salary(x, label="to"))
```

# Skills and Technology

```python
df['skills'] = df['skills'].apply(lambda x: list(x))
```

```python
def get_skill(x):
    skills = []
    for num_skills in range(len(x)):
        try:
            skills.append(x[num_skills]['name'])
        except:
            continue
    return skills

def get_level(x):
    levels = []
    for num_skills in range(len(x)):
        try:
            levels.append(x[num_skills]['level'])
        except:
            continue
    return levels
```

```python
df['technology'] = df['skills'].apply(lambda x: get_skill(x))
df['levels'] = df['skills'].apply(lambda x: get_level(x))
```
