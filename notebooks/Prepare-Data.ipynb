## Price

```python
df['employment_types'][0]
```

```python
def get_salary(data, label="from"):
    if data[0]['salary'] is not None:
        return data[0]['salary'][label]
    else:
        return data[0]['salary']
    
def get_currency(data):
    if data[0]['salary'] is not None:
        return data[0]['salary']['currency']
    else:
        return None
```

```python
df['salary_min'] = df['employment_types'].apply(lambda x: get_salary(x))
df['salary_max'] = df['employment_types'].apply(lambda x: get_salary(x, label="to"))
df['currency'] = df['employment_types'].apply(lambda x: get_currency(x))
```

## Skills and Technology

```python
df['skills'] = df['skills'].apply(lambda x: list(x))
```

```python
def get_skill(x):
    skills = []
    for num_skills in range(len(x)):
        try:
            skills.append(x[num_skills]['name'])
        except:
            continue
    return skills

def get_level(x):
    levels = []
    for num_skills in range(len(x)):
        try:
            levels.append(x[num_skills]['level'])
        except:
            continue
    return levels
```

```python
df['technology'] = df['skills'].apply(lambda x: get_skill(x))
df['levels'] = df['skills'].apply(lambda x: get_level(x))
```

```python
df.tail()
```

```python
technologies = df['marker_icon'].unique()
```

```python
technology_dict = {}
for technology in technologies:
    frameworks = []
    technology_df = df.loc[df['marker_icon'] == technology]['technology']
    for row in technology_df:
        frameworks.extend(row)
    technology_dict[technology] = list(set(frameworks))
```
